```python
import os, json, csv, pandas as pd

from selenium import webdriver

chromedriver = "chromedriver.exe"

timestampCol = ["Published"]
linkCol = ["Link"]
titleCol = ["Title"]
snippetCol = ["Snippet"]

os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(chromedriver)
csv_file = 'cbn-right.csv'

#url = cbn_links_url
with open(csv_file) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
        else:
            driver.get(row[2])
            content = driver.find_element_by_css_selector('div.field-name-body')
            
            link = row[2]
            title = driver.find_element_by_css_selector('h1.page-title').text
            timestamp = driver.find_element_by_css_selector('span.date-display-single').get_attribute('content')
            body = content.text
            
            timestampCol.append(timestamp)
            linkCol.append(link)
            titleCol.append(title)
            snippetCol.append(body)

            articles = pd.DataFrame()
            timestampCol = pd.Series(timestampCol)
            linkCol = pd.Series(linkCol)
            titleCol = pd.Series(titleCol)
            snippetCol = pd.Series(snippetCol)

            articles['Published'] = timestampCol.values
            articles['Link'] = linkCol.values
            articles['Title'] = titleCol.values
            articles['Snippets'] = snippetCol.values
            articles.to_csv('cbn-data.csv', mode='a', header=False)

            timestampCol = []
            linkCol = []
            titleCol = []
            snippetCol = []
            line_count += 1
            print('Line Count: '+str(line_count))
```
